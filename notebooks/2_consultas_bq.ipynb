# 2. Consultas e Análise Exploratória no BigQuery
Dataset: Mercado de Games

Este notebook é focado exclusivamente em se conectar ao BigQuery, rodar as queries SQL (seja nas tabelas brutas ou na View Consolidada) e trazer os resultados para um DataFrame do Pandas para construção de gráficos e análises.

## 1. Importação de Módulos

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.cloud import bigquery

# Opcional: Se rodar via Google Colab, descomente as linhas abaixo
# from google.colab import auth
# auth.authenticate_user()

pd.set_option('display.max_columns', None)
```

## 2. Autenticação e Setup

```python
# Substitua 'seu_projeto' pelo ID do seu projeto no Google Cloud
project_id = 'seu_projeto'
dataset_id = 'seu_dataset'

client = bigquery.Client(project=project_id)
```

## 3. Realizar Consultas SQL

```python
# Exemplo: Consultando a sua View consolidada
query = f'''
  SELECT *
  FROM `{project_id}.{dataset_id}.vw_games_consolidados`
  LIMIT 100
'''

# Executa a query com o client do BQ
query_job = client.query(query)
```

## 4. Converter Resultado para Pandas DataFrame

```python
# Transforma o job em um dataframe local
df_explore = query_job.to_dataframe()

# Exibir as primeiras linhas
df_explore.head()
```
