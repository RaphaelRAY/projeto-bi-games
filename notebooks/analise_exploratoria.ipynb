# Análise Exploratória de Dados
Dataset: Mercado de Games
Neste notebook, conectamos diretamente ao BigQuery (usando as credenciais definidas localmente ou no Colab) para efetuar as consultas na nossa *view* consolidada e carregar os dados num Pandas DataFrame para análise exploratória e visualização.

## 1. Importação de Módulos

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from google.cloud import bigquery

# Opcional: Descomente as linhas abaixo caso esteja rodando diretamente no Google Colab
# from google.colab import auth
# auth.authenticate_user()

pd.set_option('display.max_columns', None)
```

## 2. Autenticação e Configuração do Projeto

```python
# Substitua 'seu_projeto' pelo ID do seu projeto no Google Cloud (ex: 'projetotestemaua5584')
project_id = 'seu_projeto'

# Instancia o cliente do BigQuery
client = bigquery.Client(project=project_id)
```

## 3. Buscar Dados no BigQuery

```python
# Cria a consulta SQL para buscar os dados consolidados da View (ou de tabelas específicas)
dataset_id = 'seu_dataset' # Ajuste para o seu dataset

query = f'''
  SELECT *
  FROM `{project_id}.{dataset_id}.vw_games_consolidados`
  LIMIT 100
'''

# Executa a query
dados_bq = client.query(query)
```

## 4. Transformar em DataFrame e Explorar

```python
# Converte o resultado da consulta para um DataFrame do pandas
df_games = dados_bq.to_dataframe()

# Visualizar as primeiras linhas
df_games.head()
```
